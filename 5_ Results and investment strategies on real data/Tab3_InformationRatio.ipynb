```python
df_ir = {}

irb = np.array([compute_ir_strat_yearly(x, pd.DataFrame(dates)[0], y, np.zeros(len(x))) for x in xb_all[1:]])
df_ir['Binary'] = pd.DataFrame(irb, columns=pd.DataFrame(dates)[0].dt.year.unique()[1:], index=signal_all[1:]).round(2)
irc = np.array([compute_ir_strat_yearly(x, pd.DataFrame(dates)[0], y, np.zeros(len(x))) for x in xc_all[1:]])
df_ir['Continuous'] = pd.DataFrame(irc, columns=pd.DataFrame(dates)[0].dt.year.unique()[1:], index=signal_all[1:]).round(2)

df_ir = pd.concat(df_ir, axis=1)
df_ir
```
